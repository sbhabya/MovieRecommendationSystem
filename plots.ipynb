```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df1 = pd.read_excel("hyperparams.xlsx")
df1 = df1.set_index("rank")

df2 = pd.read_excel("rmse_vs_iters.xlsx")
df2 = df2.set_index("iterations")
df1
# print(df2.loc["rmse"])
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(1,1,1)

ax1.plot(df2.index.tolist(), df2["rmse"].tolist(), marker="x", linewidth=3)

ax1.set_ylim([0.8,0.81])
ax1.set_xticks([10,12,14,16,18,20,22,24])
ax1.set_xlabel("# of iterations in ALS optimization", fontsize=16)
ax1.set_ylabel("Validation error", fontsize=16)
ax1.set_title("Convergence in number of iterations", fontsize=14)
ax1.tick_params(labelsize=16, axis="both")
fig1.set_size_inches(6,4)
plt.savefig("iters_vs_rmse.png",dpi=320, bbox_inches="tight")
# plt.show()
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(1,1,1)

x = pd.to_numeric(df1.columns.to_series())                                                                                                                                                                                                                      .to_numpy()
y = df1.index.to_numpy()

print(f"x: {x}")
print(f"y: {y}")

z = df1.to_numpy()

xl = np.round(x, 2)
pcm = ax1.pcolormesh(xl, y, z, cmap="hot_r")
# ax1.set_xticks(x)

# ax1.set_ylim([0.8,0.81])
# ax1.set_xticks([10,12,14,16,18,20,22,24])
ax1.set_xlabel("$\lambda$ (regularizarion parameter)", fontsize=16)
ax1.set_ylabel("Rank of ALS", fontsize=16)
ax1.set_title("Error vs. hyperparameters", fontsize=14)
plt.colorbar(pcm)
# ax1.tick_params(labelsize=16, axis="both")
fig1.set_size_inches(6,4)
plt.savefig("hyperparams.png",dpi=320, bbox_inches="tight")
plt.show()
```
